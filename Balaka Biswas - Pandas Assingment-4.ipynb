```
# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]
```

Ex- 

V Kohli:

Innings:  

Total Runs: 

No. of MOM(Man of the match):

No. of boundries:

```
def info(name):
  new1 = deliveries[deliveries['batsman']==name]
  df = pd.DataFrame(new1['inning'].value_counts()).reset_index()
  inn = pd.DataFrame(new1['match_id'].value_counts())['match_id'].count()
  runs = new1['total_runs'].sum()
  bound = new1[(new1['batsman_runs']==4) | (new1['batsman_runs']==6)]['batsman_runs'].count()
  mom = matches[matches['player_of_match']==name]['player_of_match'].count()

  print("{} :\nInnings : {}\nTotal Runs : {}\nNo. of MOM(Man Of Match) : {}\nNo. of boundaries : {}".format(name,inn,runs,mom,bound))

# Test
info("V Kohli")
```

```
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```
def top(venue):
  naya = deliveries.merge(matches,left_on='match_id',right_on='id')
  naya1 = naya[naya['venue']==venue]
  return list(pd.DataFrame(naya1.groupby('batsman')['batsman_runs'].agg('sum').sort_values(ascending=False).head(4)).index)
  
# Test
top("Rajiv Gandhi International Stadium, Uppal")
```
