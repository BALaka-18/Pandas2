## Q3. Make a function to find top bowler(in terms of most wickets) against any team. Take input as team.

```python
def top_bowl(team):
  fun = delv[delv['batting_team']==team]
  return fun.groupby('bowler')['player_dismissed'].agg('count').sort_values(ascending=False).head(1)

# Test
top_bowl('Kolkata Knight Riders')
```

## Q4. Write a function which will take team name as input and return it's record of no. occurance in finals and no. of times won.  [Difficult]

ex.-- if team is CSK-- 
CSK in finals-

Played- 8
Won- 3

If any team which has not played in finals like Gujrat Lions

Gujrat Lions in finals-

Played-0


```python
def finals(team):
  df = matches.drop_duplicates('season',keep='last')
  newp = df[(df['team1'] == team) | (df['team2'] == team)]
  if newp.shape[0] != 0:
    played = newp['season'].count()
    newp_win = newp[newp['winner']==team]
    won = newp_win['season'].count()
  else:
    played = 0
  if played == 0:
    print("{} in finals-".format(team))
    print("Played-0")
  else:
    print("{} in finals-".format(team))
    print("Played-{}  Won-{}".format(played,won))

# Test
finals("Chennai Super Kings")
```

## Q5. Write a function which will take a input as batsman and in output it gives total runs scored against each team. [Medium]




```python
delv
```

```python
def bat(name):
  b = delv[delv['batsman']==name]
  return pd.DataFrame(b.groupby('bowling_team')['total_runs'].agg('sum')).reset_index()

# Test
bat('S Dhawan')
```
